````python
# voy a usar 'gemini-2.5-flash' para esta tarea
#
model = genai.GenerativeModel('gemini-2.5-flash')
````

````python
prompt = """Podrías detectar el genero de los siguientes nombres:\n
text: "Jose Maria, Maria Jose, Luis Maria, Diego Armando"
label: "M, F"
fomato de salida: json o diccionarios
"""

response = model.generate_content(contents=prompt).text
print(response)
````

````python
def parse_response_to_dict(response_string):
  """
  Converts a string containing a JSON object into a Python dictionary.

  Args:
    response_string: The input string containing the JSON object.

  Returns:
    A Python dictionary if the parsing is successful, None otherwise.
  """
  try:
    response_dict = json.loads(response_string.strip('```json\n').strip('\n```'))
    return response_dict
  except json.JSONDecodeError as e:
    print(f"Error decoding JSON: {e}")
    print(f"Raw response: {response_string}")
    return None

# Example usage:
# response_dict = parse_response_to_dict(response)
# if response_dict:
#     print(response_dict)
````
